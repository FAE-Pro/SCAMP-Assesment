```python
from functools import lru_cache

@lru_cache(maxsize = 1000)

def fibonacci(n):
    if n== 1:
        return 1
    elif n == 2:
        return 1
    elif n > 2:
        return fibonacci(n-1) + fibonacci(n-2)

for n in range(1, int(input("enter the number of fibonacci sequence you'd like to see: >> "))):
    print(n, "=", fibonacci(n))
```
